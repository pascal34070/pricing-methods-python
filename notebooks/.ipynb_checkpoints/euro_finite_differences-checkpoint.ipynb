## European option pricing by finite differences

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

from pricing_python.diffusion import BlackScholesProcess
from pricing_python.european import eur_bs_analytical_price 

sigma = 0.2
riskfree = 0.05
maturity = 1
S0 = 100
K = 110
model = BlackScholesProcess(sigma=sigma)
```

### Solve the BS equation : explicit, implicit and Crank-nicolson schemes

```python
from pricing_python.european import eur_fd_price

x_step = 1000
time_step = 1000
S_max = 2*K

strtime = datetime.now()
_, S_imp, grid_vals_implicit = eur_fd_price(model, S0, K, riskfree,
                                            maturity, call=True, method='implicit',
                                            x_step=x_step, time_step=time_step, S_max=S_max)
endtime = datetime.now()
imp_exectime = (endtime - strtime).total_seconds()

strtime = datetime.now()
_, S_exp, grid_vals_explicit = eur_fd_price(model, S0, K, riskfree,
                                            maturity, call=True, method='explicit',
                                            x_step=100, time_step=time_step, S_max=S_max)
endtime = datetime.now()
exp_exectime = (endtime - strtime).total_seconds()

strtime = datetime.now()
_, S_cn, grid_vals_cn = eur_fd_price(model, S0, K, riskfree,
                               maturity, call=True, method='cn',
                               x_step=x_step, time_step=time_step, S_max=S_max,
                               theta=0.5)
endtime = datetime.now()
cn_exectime = (endtime - strtime).total_seconds()

# interpolate the explicit grid to implicit grid
dt = maturity/(time_step - 1)

npv_exp = np.zeros((S_imp.shape[0], time_step))
for n in range(time_step):
    npv_exp[:,n] = np.interp(S_imp, S_exp, grid_vals_explicit[:, n])
```

### Solve analytically 

```python

npv_analytical = np.zeros((S_imp.shape[0], time_step))
for i in range(len(S_imp)):
    for n in range(time_step):
        npv_analytical[i, n] = eur_bs_analytical_price(S_imp[i], K, riskfree, sigma, maturity-n*dt, call=True)
```

```python
mae_exp = np.mean(np.abs(npv_exp-npv_analytical))
mae_imp = np.mean(np.abs(grid_vals_implicit-npv_analytical))
mae_cn = np.mean(np.abs(grid_vals_cn-npv_analytical))
```

### Plotting

```python

textstr = '\n'.join((
    r'Implicit: MAE {},  Solving exec. {}s'.format(np.round(mae_imp, 5), np.round(imp_exectime, 4)),
    r'Explicit: MAE {},  Solving exec. {}s'.format(np.round(mae_exp, 4), np.round(exp_exectime, 4)),
    r'CN:     MAE {},  Solving exec. {}s'.format(np.round(mae_cn, 6), np.round(cn_exectime, 4))))

props = dict(boxstyle='round', facecolor='white', alpha=0.5)

plt.figure()
plt.figure(figsize = (10,6))
plt.plot(S_imp, npv_analytical[:,0], color='black', label='BS Analytical')
plt.plot(S_imp, grid_vals_implicit[:,0], label='Implicit scheme')
plt.plot(S_imp, npv_exp[:,0], label='Explicit scheme')
plt.plot(S_imp, grid_vals_cn[:,0], label='Crank-Nicolson')
plt.text(0.0, 80., textstr, fontsize=14,
        verticalalignment='top', bbox=props)
plt.legend(prop={"size":13})
plt.title("Call option values at t=0", fontsize=14)
plt.xlabel("Underlying price", fontsize=14)
plt.ylabel("Option value", fontsize=14)
plt.show()
```
