## European basket option by Deep Learning PDE solving

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from datetime import datetime
from scipy.stats import random_correlation
from scipy.stats.mstats import gmean
from pricing_python.diffusion import BlackScholesProcessMulti

# Option parameters 
dimension = 10
r = 0.05           # Interest rate
K = 100            # Strike
T = 1              # Terminal time
init_vals = dimension*[100.]      # Initial price

sigmas = np.array(dimension*[0.2])
correlations = 0.25 * np.ones((dimension, dimension))
np.fill_diagonal(correlations, 1)
#correlations = np.eye(dimension)
print(correlations)
model = BlackScholesProcessMulti(sigmas=sigmas, correlations=correlations)

def payoff_call_basket(S, K):
    """
    payoff call basket option
    :param S (array-like): price matrix of dim dxN
    :param K (float): strike
    :return: array of dim N
    """
    G = np.maximum(np.mean(S, axis=1) - K, 0)
    return G

def payoff_call_basket_gmean(S, K):
    """
    payoff call basket option
    :param S (array-like): price matrix of dim dxN
    :param K (float): strike
    :return: array of dim N
    """
    G = np.maximum(gmean(S, axis=1) - K, 0)
    return G
```

### Monte-Carlo price

```python
from pricing_python.eur_basket import eur_basket_mc

mc_price, _, err = eur_basket_mc(model,n_path=100000, init_val=init_vals, K=K, riskfree=r,
                                 maturity=T, payoff_fun=payoff_call_basket)

print('price: ', mc_price, 'conf inf: ', mc_price - err, 'conf sup: ', mc_price + err)
```

### Deep Galerkin Method

```python
from pricing_python.dgm.eur_basket import EurBasketDGM
from pricing_python.dgm.dgm import DGMNet
import tensorflow as tf

dgm_net = DGMNet(hidden_units=50, n_layers=3, input_dim=model.dim,
                 lstm_activation_1="tanh", lstm_activation_2="tanh",
                 last_activation="linear")

dgm_pricer = EurBasketDGM(model, K, maturity=T, riskfree=r, dgm_net=dgm_net,
                          S_max=2*K, n_sim_interior=2000, n_sim_terminal=200)

lr_schedule = tf.keras.optimizers.schedules.PiecewiseConstantDecay(
                boundaries=[200], values=[1., 0.1])

dgm_pricer.solve(sampling_iter=5, steps_per_sample=20, learning_rate=.001)
```

```python
# evaluate the model at the initial values
S_val = np.array(init_vals).reshape(1,-1)
t_val = np.array([[dgm_pricer.t_min]])
dgm_pricer.eval(S_val, t_val).numpy()
```

### Deep BSDE 

```python
from pricing_python.deep_bsde.networks import mlp
from pricing_python.deep_bsde.eur_basket import EurBasketBSDE

net_y = mlp(model.dim, output_dim=1, hidden_shape=(model.dim+10,model.dim+10),
            activation='relu', use_bias=True,
            batch_normalization=True)

net_z = mlp(model.dim, output_dim=model.dim, hidden_shape=(model.dim+10,model.dim+10),
            activation='relu', use_bias=True,
            batch_normalization=True)

bsde_pricer = EurBasketBSDE(model, init_vals, K, maturity=T, riskfree=r,
                            network_y=net_y, network_z=net_z,
                            n_path=10000, n_step=50)
```

```python
bsde_pricer.solve(epochs=50, batch_size=500, learning_rate=0.001)
```

### Deep BSDE Connected network mini-batch GD

```python
from pricing_python.deep_bsde.eur_basket import EurBasketBSDEConnectedNetV2, payoff_call_mean

bsde_pricer = EurBasketBSDEConnectedNetV2(model, init_vals, K,
                                          maturity=T, riskfree=r, y_init_range=[6,8],
                                          payoff_fn=payoff_call_mean, n_step=100,
                                          hidden_shape=(model.dim+10,model.dim+10),
                                          activation='relu', use_bias=False)
```

```python
import tensorflow as tf
lr_schedule = tf.keras.optimizers.schedules.PiecewiseConstantDecay(boundaries=[3000], values=[.005, .001])

strttime = datetime.now()
bsde_pricer.solve(sampling_iter=4000, batch_size=64, learning_rate=lr_schedule)
endtime = datetime.now()
print('solving time: ', endtime - strttime)
```

```python
iterations = np.arange(0,4000,50)
lcurve = np.array(bsde_pricer.train_losses)[:,0]
y_0 = np.array(bsde_pricer.train_losses)[:,1]

fig, ax1 = plt.subplots(figsize = (10,6))

ax1.plot(iterations, lcurve, label='loss', color='darkred')
ax1.set_xlabel("Iterations", fontsize=14)
ax1.set_ylabel("Loss", fontsize=14, color='darkred')
# ax1.legend(prop={"size":13})
ax2 = ax1.twinx()
ax2.plot(iterations, y_0, label='Y_0')
ax2.set_ylabel("Y_0 value", fontsize=14)
ax2.legend(prop={"size":13})

plt.title("Learning curve", fontsize=14)
plt.grid()
plt.show()
```
